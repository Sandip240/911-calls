```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("911.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head(5)
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x=df['Reason'],data=df)
plt.show()
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour']=df['timeStamp'].apply(lambda time:time.hour)
df['Month']=df['timeStamp'].apply(lambda time:time.month)
df['Day of Week']=df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week']=df['Day of Week'].map(dmap)
```

```python
sns.countplot(x=df['Day of Week'],data=df,hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['twp'].plot(color='blue',ls='--')
plt.show()
```

```python
rmonth=byMonth.reset_index()
rmonth['Month']=pd.Categorical(rmonth['Month'],categories=['jan','feb','mar','apr','may','jun','july','aug','sep','oct','nov','des'],ordered=True)
rmonth['Month']=rmonth['Month'].cat.codes + 1
sns.lmplot(x='Month',y='twp',data=rmonth)
plt.show()
```

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.legend(loc=0)
plt.show()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
plt.show()

```

```python
df[df['Reason']=='EMS'].groupby('Day of Week').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
plt.show()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
plt.show()
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(data=dayHour,cmap='viridis')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.clustermap(data=dayHour)
plt.show()
```

```python
daymonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
daymonth.head()
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(data=daymonth,cmap='viridis')
plt.show()
```

```python

sns.clustermap(data=daymonth,cmap='viridis')
plt.show()
```
